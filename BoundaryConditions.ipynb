```python
import matplotlib.pyplot as plt
import math

n = 10
h = 1 / n

def origin(x):
  return math.exp(x) + math.exp(-x) + 3.3 * x * x - 3.3 * x - 2

def p(x):
  return 1

def q(x):
  return 8.6 + 3.3 * x * (1 - x)

def A():
  return 2 + h * h

def B(x):
  return q(x) * h * h

l = [0]
m = [0]
x_values = [0]

for i in range (1, n + 1):
  x_values.append(x_values[-1] + h)
  l.append(1 / (A() - l[i - 1]))
  m.append((m[i - 1] - B(x_values[i - 1])) / (A() - l[i - 1]))

y_values = [((2 * math.e + 1.3)*h + m[n]) / (1 + h - l[n])]

for i in range (n - 1, 0, -1):
  y_values.append(l[i + 1] * y_values[n - 1 - i] + m[i + 1])

y_values.append(0)

y_values.reverse();

origins = []

for i in range (0, n + 1):
  origins.append(origin(x_values[i]))
```

```python
plt.grid()
plt.plot(x_values, y_values, '-bo')
plt.show()
```

```python
plt.grid()
plt.plot(x_values, origins, '-go')
plt.show()
```
